The aim of this project is to clean the data and analyze the included used car listings. You'll also become familiar with some of the unique benefits jupyter notebook provides for pandas.

Let's start by importing the libraries we need and reading the dataset into pandas.

<h2>Data Exploration and Data Cleaning</h2>
Let's look for text columns where most values are the same as this can often be dropped since it does not provide very useful info. Also, we will identify numeric data sotred as string and clean and convert that.

```python
autos.describe(include='all') #all to get categorical as well as numeric data
```

The seller column stands out as one to drop because it only has 2 unique values out of 50,000! Same for offer_type and ab_test and gearbox, and unrepaired_damage. Actually, almost all of these are binary choices so we can probably leave them after inspection except possibly for offer_type.

Vehicle and gearbox are missing data in about 10% of the rows so we should investigate this. Same for model, fuel_type and unrepaired_damage.

The numeric columns which need to be cleaned and converted include: price, odometer and num_photos.

First the numeric columns:

```python
print(autos["num_photos"].value_counts())
print(autos["seller"].value_counts())
autos["offer_type"].value_counts()
```

num_photos has all zeros so we can drop this one. Same for seller and offer_type which are both essentially a single value.

```python
autos.drop(["num_photos", "seller", "offer_type"], axis=1)
```

```python
autos["unrepaired_damage"].value_counts()
```

There is some distribution among yes and no values in unrepaired_damage so we will retain this column

Now let's remove non-numeric characters from the price and odometer columns then convert to numeric and rename the column.

```python
print(autos["price"].value_counts())
autos["price"] = autos["price"].str.replace("$", "").str.replace(",", "").astype(int)
autos["price"].head()
```

```python
print(autos["odometer"].value_counts())
autos["odometer"] = (autos["odometer"]
                        .str.replace(",","")
                        .str.replace("km","")
                        .astype(int))
autos.rename({"odometer":"odometer_km"}, inplace=True, axis=1)
autos["odometer_km"].head()
```

<h2>Explore Price and Odometer</h2>

```python
print(autos["price"].unique().shape)
print(autos["price"].describe())
print(autos["price"].value_counts().head(15))
autos["price"].value_counts().sort_index(ascending=False).head(15)
```

The average price is about 9,840 but we have many zero prices and several north of 1,000,000. It's unlikely to have cars over 1,000,000 on an ebay site so let's remove those and the zeros.

Above we removed zero prices and abnormally large prices above 350,000. We can see the mean came down to 5899.

Above we can see that 150,000 has the most common mileage in the dataset and higher mileage cars dominate. The mean is 125,733 km. We do not see any outliers worth dropping here.

The `registration_year` column above has some outlier values with a max of 9999 which is not a valid year, and a min of 1000.

```python
print(autos["registration_year"].sort_values().head(20))
print(autos["registration_year"].sort_values().tail(30))
autos[autos["registration_year"].between(1910, 2016)]["registration_year"].describe()
```

It would make sense, per above, to throw out years prior to 1910 and years after 2016.

Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s.

Once we remove the outliers as seen above we can see normal descriptive statistics of this column.

<h2>Exploring Price by Brand</h2>

```python
autos["brand"].value_counts(normalize=True)
```

Volkswagen is by far the most popular brand with more than 21% of the values. We will aggregate the brands which each represent at least 5% of the total counts in the `brand` column.

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > 0.05].index
common_brands
```

```python
brands_mean_prices = {}
for brand in common_brands:
    brand_table = autos[autos["brand"] == brand]
    mean_price = brand_table["price"].mean()
    brands_mean_prices[brand] = mean_price
print(brands_mean_prices)
```

Above we see the mean prices for the common brands of cars listed on eBay. Mercedez is the highest by far with Opel being the lowest.

<h2>Exploring mileage using the Pandas Dataframe and Series Constructors</h2>

```python
brands_mean_mileage = {}
for brand in common_brands:
    brand_table = autos[autos["brand"] == brand]
    mean_mileage = brand_table["odometer_km"].mean()
    brands_mean_mileage[brand] = mean_mileage
print(brands_mean_mileage)

#convert dicts to series objects
price_series = pd.Series(brands_mean_prices)
print(price_series)
mileage_series = pd.Series(brands_mean_mileage)
print(mileage_series)

#convert price series to dataframe
df = pd.DataFrame(price_series, columns=['mean_price'])
df

#Assign mileage series to dataframe
df['mean_mileage'] = mileage_series
df
```

The new DataFrame is displayed above. The mean mileages are pretty uniformly distributed so the mileage does not appear to have much of an effect on the mean price differences among brands.
