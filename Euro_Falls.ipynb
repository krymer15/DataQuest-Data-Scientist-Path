<h2>Introducing the Dataset</h2>

```python
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates
```

```python
exchange_rates.info()
```

Several columns are string types and only a few are actually float types for the rates. There are also some columns with null values. There are 5699 rows and 41 columns.

Do some data cleaning on the US Dollar column.

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates
```

```python
exchange_rates.info()
```

Isolate Time and US_Dollar columns:

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

We have 62 values with `-`.

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != "-"]
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

<h2>Rolling Mean (aka Moving Average)</h2>

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# enable Jupyter to display graphs

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
rolling_mean = euro_to_dollar['US_dollar'].rolling(30).mean()
rolling_mean.tail()
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'].rolling(30).mean())
plt.show()
```

<h2>Storytelling Data Visualization: Euro/USD Hits 12-Year Low</h2>

Over the period of 2014 and 2015 the EUR/USD pair went off a cliff nosediving to a 12-year low against the dollar in 2015. Three factors contributed to this fall: (1) relatively strong US economy, (2) weaker Euro economy and (3) struggles for the Eurozone to exit its sovereign debt crisis.

```python
# Create 2014-2015 EUR/USD dataframe
eurofall = (euro_to_dollar[(euro_to_dollar['Time'].dt.year >= 2014) & 
                           (euro_to_dollar['Time'].dt.year <= 2015)].copy())
eurofall
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')
```

```python
# add plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(eurofall['Time'], eurofall['US_dollar'].rolling(30).mean())

### Highlihting the drop
ax.axvspan(xmin=735390, xmax=735700, ymin=0, ymax=1, alpha=0.5, color='grey')

# Modifying tick labels
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 735300
for qtr in ['Q2 14', 'Q3 14', 'Q4 14', 'Q1 15', 'Q2 15', 'Q3 15', 'Q4 15', 'Q1 16']:
    ax.text(x, 1.04, qtr, alpha=0.5, fontsize=11)
    x += 91.67
    
y = 1.097
for rate in ['1.10', '1.15', '1.20', '1.25', '1.30', '1.35']:
    ax.text(735190, y, rate, alpha=0.5, fontsize=11)
    y += 0.05
    
# add title and subtitle
ax.text(735310, 1.45, 'EUR/USD Falls While Trying to Exit Debt Crisis', weight='bold')
ax.text(735200, 1.43, 'Stronger US Economy, Weaker Eurozone Amid Struggles to Exit Debt Crisis')

# Grid transparency
ax.grid(alpha=0.35)

### Adding a signature
ax.text(735200.0, 1.02, '©DATAQUEST' + ' '*94 + 'Source: European Central Bank',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=10)

# show plot
plt.show()
```
